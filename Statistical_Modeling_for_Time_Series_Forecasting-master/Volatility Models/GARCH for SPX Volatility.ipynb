```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.graphics.tsaplots as sgt
from statsmodels.tsa.stattools import adfuller, kpss

from sklearn.metrics import mean_squared_error

from arch import arch_model

import warnings

sns.set()
warnings.filterwarnings("ignore")
```

# Importing Dataset

```python
file_path = "../data.csv"
```

```python
data_backup = pd.read_csv(file_path)
data_backup.Date = pd.to_datetime(data_backup.Date)
data_backup.set_index("Date", inplace = True)
```

```python
data = data_backup[["spx_vol", "spx_ret"]]
```

# Train Test Split

```python
train_df = data.loc[:"2018-12-31"]
test_df = data.loc["2019-01-01":]
```

# GARCH Model

$Var(y_t|y_{t-1}) = \sigma^{2}_t = \omega + (\alpha_1\epsilon^{2}_{t-1} + \alpha_2\epsilon^{2}_{t-2} + ... + \alpha_p\epsilon^{2}_{t-p}) + (\beta_1\sigma^{2}_{t-1} + \beta_2\sigma^{2}_{t-2} + ... + \beta_q\sigma^{2}_{t-q})$

where, 
- $y_t$: Time Series at time $t$
- $\epsilon_t$: Residuals at time $t$ 
- $\sigma_t$: Volatility of Time Series at time $t$.
- $\omega$: Constant Term
- $\alpha$: Coefficients for the $\epsilon$ terms
- $\beta$: Coefficients for the $\sigma$ terms.
- $p$: Number of lagged $\epsilon$ terms
- $q$: Number of lagged $\sigma$ terms

## Estimating Signitficant Lags

```python
plt.rcParams["figure.figsize"] = 18, 5

fig, axes = plt.subplots(1, 2)

sgt.plot_pacf(train_df.spx_ret[1:], zero = False, lags = 40, ax = axes[0])
# sgt.plot_pacf(train_df.spx_ret[1:].abs(), zero = False, lags = 40, ax = axes[1])
sgt.plot_pacf(train_df.spx_vol[1:], zero = False, lags = 40, ax = axes[1])

axes[0].set_title("PACF For Returns", size = 24)
axes[1].set_title("PACF For Squared Returns (Volatility)", size = 24)

plt.show()
```

Looking at the **PACF** plots for returns and Volatility, it is clear that the first 2 lags are very significant and then the significance levels drop significantly. So we will build a **GARCH(2, 2)** Model.

## Building a GARCH(2, 2) Model

**Data Preperation**

```python
garch_df = pd.DataFrame(data["spx_ret"].shift(1).loc[data.index])
garch_df.at[train_df.index, "spx_ret"] = train_df["spx_ret"]
```

**Model Instantiation**

```python
model = arch_model(garch_df["spx_ret"][1:], p = 2, q = 2, vol = "GARCH")
model_results = model.fit(last_obs = test_df.index[0], update_freq = 5)
```

**Model Summary**

```python
model_results.summary()
```

**Model Predictions**

```python
predictions_df = test_df.copy()
predictions_df["Predictions"] = model_results.forecast().residual_variance.loc[test_df.index]
```

**Plotting Predictions**

```python
plt.rcParams["figure.figsize"] = 18, 5

plt.plot(predictions_df["spx_ret"].abs())
plt.plot(predictions_df["Predictions"])

plt.title("Actuals vs Predictions for SPX Volatility", size = 24)

plt.show()
```

**RMSE**

```python
np.sqrt(mean_squared_error(predictions_df["spx_ret"].abs(), predictions_df["Predictions"]))
```
